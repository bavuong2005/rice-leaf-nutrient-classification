```python
import gc
import time
import traceback
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy
from torch.utils.data import DataLoader, Subset
from torchvision import transforms, models
from torchvision.datasets import ImageFolder
from sklearn.model_selection import StratifiedKFold, StratifiedShuffleSplit
from sklearn.metrics import classification_report

# ==============================================================================
# 1. SETUP MÔI TRƯỜNG & HÀM BỔ TRỢ
# ==============================================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# --- Hàm dọn dẹp bộ nhớ ---
def clear_cuda():
    gc.collect()
    if torch.cuda.is_available():
        try:
            torch.cuda.empty_cache()
        except Exception:
            pass

# --- Hàm tạo Model Xception (Có fallback) ---
try:
    import timm
    def create_pretrained_xception(num_classes=3):
        model = timm.create_model('xception', pretrained=True, num_classes=num_classes)
        return model
    print("✓ Using timm Xception")
except ImportError:
    def create_pretrained_xception(num_classes=3):
        model = models.efficientnet_b0(weights='DEFAULT')
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
        return model
    print("⚠ timm not found, using EfficientNet-B0 fallback")

# --- Hàm tạo DataLoader an toàn ---
def make_loader(dataset, batch_size, shuffle):
    # num_workers=2, pin_memory=True để tối ưu tốc độ load data lên GPU
    return DataLoader(dataset, batch_size=batch_size, shuffle=shuffle, 
                      num_workers=2, pin_memory=True)

# --- Class EarlyStopping ---
class EarlyStopping:
    def __init__(self, patience=12, min_delta=0.0):
        self.patience = patience
        self.min_delta = min_delta
        self.best_loss = None
        self.counter = 0
        self.stop = False

    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0

# ==============================================================================
# 2. LOAD DATASET (Đảm bảo test_ds được định nghĩa)
# ==============================================================================
# Cấu hình
BATCH_SIZE = 32
NUM_EPOCHS = 100
LR = 0.001
N_SPLITS = 5
IMG_SIZE = (1024, 256)
DATA_ROOT = "/kaggle/input/nutrientdeficiencysymptomsinrice/rice_plant_lacks_nutrients/"

# Transforms
train_transform = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(20),
    transforms.RandomAffine(degrees=0, scale=(0.8, 1.2)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

eval_transform = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Load Data
print("\nLoading dataset...")
ds_full_train = ImageFolder(DATA_ROOT, transform=train_transform)
ds_full_eval = ImageFolder(DATA_ROOT, transform=eval_transform)
labels = np.array(ds_full_eval.targets)
print(f"✓ Found {len(labels)} images.")

# Chia Train/Test (Định nghĩa test_ds ở đây)
splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=42)
trainval_idx, test_idx = next(splitter.split(np.zeros(len(labels)), labels))

test_ds = Subset(ds_full_eval, test_idx)
trainval_labels = labels[trainval_idx]

print(f"✓ Train+Val: {len(trainval_idx)} | Test: {len(test_idx)}")

# ==============================================================================
# 3. TRAINING LOOP (K-FOLD)
# ==============================================================================
use_amp = (str(device) == "cuda")
scaler = torch.amp.GradScaler('cuda') if use_amp else None
skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

# Tạo Test Loader (Đã có test_ds)
test_loader = make_loader(test_ds, batch_size=BATCH_SIZE, shuffle=False)

results = []
best_overall_acc = 0
best_overall_state = None
best_overall_fold = None
fold = 1

print(f"\n{'='*60}")
print(f"STARTING TRAINING | BS={BATCH_SIZE} | Img={IMG_SIZE} | GradCheckpointing=ON")
print(f"{'='*60}")

for train_idx, val_idx in skf.split(np.zeros(len(trainval_labels)), trainval_labels):
    print(f"\n{'='*20} FOLD {fold}/{N_SPLITS} {'='*20}")

    train_subset = Subset(ds_full_train, trainval_idx[train_idx])
    val_subset   = Subset(ds_full_eval,  trainval_idx[val_idx])

    # Reset Model & Clean GPU
    clear_cuda()
    model = create_pretrained_xception(num_classes=len(ds_full_eval.classes))
    
    # [QUAN TRỌNG] Bật Gradient Checkpointing để tránh OOM
    try:
        model.set_grad_checkpointing(True)
        print("✓ Gradient Checkpointing enabled.")
    except Exception as e:
        print(f"⚠ Warning: {e}")

    model = model.to(device)

    # Freeze strategy
    for name, p in model.named_parameters():
        p.requires_grad = False
    
    # Unfreeze 30 layers cuối
    K = 30
    all_params = list(model.named_parameters())
    if K > 0:
        for name, p in all_params[-K:]:
            p.requires_grad = True
    
    # Fallback unfreeze head
    trainable_params = list(filter(lambda p: p.requires_grad, model.parameters()))
    if not trainable_params:
        for name, p in model.named_parameters():
            if 'classifier' in name or 'head' in name or 'fc' in name:
                p.requires_grad = True
        trainable_params = list(filter(lambda p: p.requires_grad, model.parameters()))

    # Optimizer
    opt = optim.AdamW(trainable_params, lr=LR, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(opt, mode='min', factor=0.1, patience=5)
    criterion = nn.CrossEntropyLoss()
    earlystop = EarlyStopping(patience=12)

    # Loaders
    train_loader = make_loader(train_subset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader   = make_loader(val_subset,   batch_size=BATCH_SIZE, shuffle=False)

    # Metrics containers
    train_losses, val_losses = [], []
    train_accs, val_accs = [], []  # ✅ Thêm tracking accuracy
    best_val_loss = float("inf")
    best_state = None
    train_times = []  # ✅ Lưu thời gian train mỗi epoch

    try:
        for epoch in range(NUM_EPOCHS):
            # ========== TRAIN ==========
            train_start_time = time.time()  # ✅ Bắt đầu đo thời gian train
            
            model.train()
            r_loss, correct, total = 0.0, 0, 0

            for x, y in train_loader:
                x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
                opt.zero_grad(set_to_none=True) # RAM optimization

                if use_amp:
                    with torch.amp.autocast('cuda', enabled=True):
                        out = model(x)
                        loss = criterion(out, y)
                    scaler.scale(loss).backward()
                    scaler.step(opt)
                    scaler.update()
                else:
                    out = model(x)
                    loss = criterion(out, y)
                    loss.backward()
                    opt.step()

                r_loss += loss.item()
                preds = torch.argmax(out, dim=1)
                correct += (preds == y).sum().item()
                total += y.size(0)
                
                del x, y, out, loss, preds # Delete immediately

            t_loss = r_loss / max(1, len(train_loader))
            t_acc = correct / max(1, total)
            
            train_time = time.time() - train_start_time  # ✅ Thời gian train
            train_times.append(train_time)  # ✅ Lưu lại

            # ========== VAL ==========
            val_start_time = time.time()  # ✅ Bắt đầu đo thời gian val
            
            model.eval()
            v_loss_sum, v_correct, v_total = 0.0, 0, 0
            with torch.no_grad():
                for x, y in val_loader:
                    x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
                    if use_amp:
                        with torch.amp.autocast('cuda', enabled=True):
                            out = model(x)
                            loss = criterion(out, y)
                    else:
                        out = model(x)
                        loss = criterion(out, y)
                    
                    v_loss_sum += loss.item()
                    preds = torch.argmax(out, dim=1)
                    v_correct += (preds == y).sum().item()
                    v_total += y.size(0)
                    del x, y, out, loss, preds

            v_loss = v_loss_sum / max(1, len(val_loader))
            v_acc = v_correct / max(1, v_total)
            
            val_time = time.time() - val_start_time  # ✅ Thời gian val
            total_time = train_time + val_time  # ✅ Tổng thời gian epoch
            
            # ✅ In ra cả Train Time và Total Time
            print(f"Ep {epoch+1:3d} | T_Loss:{t_loss:.4f} Acc:{t_acc:.4f} | "
                  f"V_Loss:{v_loss:.4f} Acc:{v_acc:.4f} | "
                  f"Train:{train_time:.1f}s Total:{total_time:.1f}s")
            
            train_losses.append(t_loss)
            val_losses.append(v_loss)
            train_accs.append(t_acc)  # ✅ Lưu train accuracy
            val_accs.append(v_acc)    # ✅ Lưu val accuracy

            # Scheduler & Checkpoint
            try: scheduler.step(v_loss)
            except: pass

            if v_loss < best_val_loss:
                best_val_loss = v_loss
                best_state = {k: v.cpu() for k, v in model.state_dict().items()}

            earlystop(v_loss)
            if earlystop.stop:
                print(f"⚠ Early stopping at epoch {epoch+1}")
                break

        # ✅ In thời gian train trung bình mỗi epoch
        avg_train_time = np.mean(train_times) if len(train_times) > 0 else 0
        print(f"\n⏱️ Average Train Time per Epoch (Fold {fold}): {avg_train_time:.2f}s")

        # ✅ VẼ BIỂU ĐỒ TRAIN/VAL CHO FOLD
        try:
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
            
            # Plot Loss
            ax1.plot(train_losses, label="Train Loss", linewidth=2, color='#1F77B4')
            ax1.plot(val_losses, label="Val Loss", linewidth=2, color='#FF7F0E')
            ax1.set_title(f"Fold {fold} - Loss Curve", fontsize=14, fontweight='bold')
            ax1.set_xlabel("Epoch", fontsize=12)
            ax1.set_ylabel("Loss", fontsize=12)
            ax1.legend(fontsize=11)
            ax1.grid(True, alpha=0.3, linestyle='--')
            
            # Plot Accuracy
            ax2.plot(train_accs, label="Train Acc", linewidth=2, color='#1F77B4')
            ax2.plot(val_accs, label="Val Acc", linewidth=2, color='#FF7F0E')
            ax2.set_title(f"Fold {fold} - Accuracy Curve", fontsize=14, fontweight='bold')
            ax2.set_xlabel("Epoch", fontsize=12)
            ax2.set_ylabel("Accuracy", fontsize=12)
            ax2.legend(fontsize=11)
            ax2.grid(True, alpha=0.3, linestyle='--')
            
            plt.tight_layout()
            plt.show()
            plt.close(fig)  # Đóng figure để giải phóng bộ nhớ
        except Exception as plot_err:
            print(f"⚠ Plot error: {plot_err}")

        # --- EVALUATE FOLD ---
        if best_state:
            model.load_state_dict(best_state)
        model.to(device).eval()
        
        y_true, y_pred = [], []
        with torch.no_grad():
            for x, y in test_loader:
                x = x.to(device, non_blocking=True)
                if use_amp:
                    with torch.amp.autocast('cuda', enabled=True):
                        out = model(x)
                else:
                    out = model(x)
                y_pred.extend(torch.argmax(out, 1).cpu().numpy())
                y_true.extend(y.numpy())
                del x, out

        acc = np.mean(np.array(y_true) == np.array(y_pred))
        print(f"\n>>> Fold {fold} Test Acc: {acc:.4f}")
        print(classification_report(y_true, y_pred, target_names=ds_full_eval.classes, zero_division=0))
        
        results.append(acc)
        
        if acc > best_overall_acc:
            best_overall_acc = acc
            best_overall_state = best_state
            best_overall_fold = fold

    except RuntimeError as e:
        print(f"❌ ERROR FOLD {fold}: {e}")
        traceback.print_exc()
        results.append(0.0)
    finally:
        del model, opt
        clear_cuda()
    
    fold += 1

print(f"\n{'='*60}")
print(f"TRAINING COMPLETED")
print(f"{'='*60}")
print(f"Results per fold: {results}")
print(f"Final Average Accuracy: {np.mean(results):.4f}")
```

```python
save_path = "best_pretrained_xception_overall.pth"
torch.save(best_overall_state, save_path)
print(f"✓ Best model saved to: {save_path}")
```

```python
# --- COPY TOÀN BỘ VÀO CELL 3 ---
avg_acc = np.mean(results)
std_acc = np.std(results)
miss_rate = 1 - avg_acc

print(f"\n{'='*60}")
print("FINAL RESULTS")
print(f"{'='*60}")
print(f"\n5-Fold Cross-Validation Results:")
for i, acc in enumerate(results, 1):
    print(f"  Fold {i}: {acc:.4f}")

print(f"\n{'='*40}")
print(f"Average Accuracy: {avg_acc:.4f} ± {std_acc:.4f}")
print(f"Miss-classification Rate: {miss_rate:.4f}")
print(f"Best Fold: {best_overall_fold} with Accuracy: {best_overall_acc:.4f}")
print(f"{'='*40}")

# --- PHẦN ĐÃ SỬA LẠI LOGIC ĐẾM THAM SỐ ---
print(f"\nModel Parameters Information:")

# 1. Tạo model mới để kiểm tra
test_model = create_pretrained_xception(num_classes=len(ds_full_eval.classes))

# 2. === QUAN TRỌNG: ÁP DỤNG LẠI LOGIC FREEZE (để đếm đúng) ===
# Freeze toàn bộ
for p in test_model.parameters():
    p.requires_grad = False

# Mở khóa K tensor cuối (Giống hệt biến K trong vòng lặp train, ở code cũ là 30)
K = 30 
all_params = list(test_model.named_parameters())
if K > 0:
    for name, p in all_params[-K:]:
        p.requires_grad = True
# ==============================================================

# 3. Tính toán và in ra
model_params = sum(p.numel() for p in test_model.parameters())
trainable_params = sum(p.numel() for p in test_model.parameters() if p.requires_grad)

print(f"  Total: {model_params:,} ({model_params/1e6:.2f}M)")
print(f"  Trainable: {trainable_params:,} ({trainable_params/1e6:.2f}M)")
print(f"  (Note: Đã tính theo cấu hình Freeze backbone + Unfreeze {K} tensors cuối)")

del test_model
```

```python
print(f"\n{'='*60}")
print("DETAILED TEST SET ANALYSIS (BEST MODEL)")
print(f"{'='*60}")

# Load best model
model = create_pretrained_xception(num_classes=len(ds_full_eval.classes)).to(device)
model.load_state_dict(best_overall_state)
model.eval()

# Get predictions with probabilities and store images
y_true, y_pred, y_probs = [], [], []
test_images = []  # Lưu ảnh để visualize sau

with torch.no_grad():
    for x, y in test_loader:
        # Lưu ảnh gốc (denormalize để hiển thị)
        for img in x:
            test_images.append(img.cpu())
        
        x = x.to(device)
        if use_amp:
            with torch.amp.autocast("cuda"):
                out = model(x)
        else:
            out = model(x)
        
        probs = torch.softmax(out, dim=1)
        pred = torch.argmax(out, dim=1).cpu().numpy()
        
        y_pred.extend(pred)
        y_true.extend(y.numpy())
        y_probs.extend(probs.cpu().numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)
y_probs = np.array(y_probs)

# Find wrong predictions
wrong_indices = np.where(y_true != y_pred)[0]

print(f"\nTest Set Summary:")
print(f"  Total samples: {len(y_true)}")
print(f"  Correct predictions: {np.sum(y_true == y_pred)} ({np.sum(y_true == y_pred)/len(y_true)*100:.2f}%)")
print(f"  Wrong predictions: {len(wrong_indices)} ({len(wrong_indices)/len(y_true)*100:.2f}%)")

if len(wrong_indices) > 0:
    print(f"\n{'='*70}")
    print("WRONG PREDICTIONS DETAILS:")
    print(f"{'='*70}")
    print(f"{'Index':<8} {'True Label':<20} {'Predicted':<20} {'Confidence':<12}")
    print("-" * 70)
    
    for idx in wrong_indices[:30]:
        true_label = ds_full_eval.classes[y_true[idx]]
        pred_label = ds_full_eval.classes[y_pred[idx]]
        confidence = y_probs[idx][y_pred[idx]] * 100
        print(f"{idx:<8} {true_label:<20} {pred_label:<20} {confidence:>10.2f}%")
    
    if len(wrong_indices) > 30:
        print(f"\n... and {len(wrong_indices) - 30} more wrong predictions")
    
    # ✅ VISUALIZE WRONG PREDICTIONS
    print(f"\n{'='*60}")
    print("VISUALIZING WRONG PREDICTIONS")
    print(f"{'='*60}")
    
    # Denormalize function
    def denormalize(img_tensor):
        mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
        std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
        img = img_tensor * std + mean
        img = torch.clamp(img, 0, 1)
        return img.permute(1, 2, 0).numpy()
    
    # Hiển thị tối đa 20 ảnh sai đầu tiên
    num_show = min(20, len(wrong_indices))
    cols = 5
    rows = (num_show + cols - 1) // cols
    
    fig, axes = plt.subplots(rows, cols, figsize=(20, rows * 4))
    axes = axes.flatten() if num_show > 1 else [axes]
    
    for i, idx in enumerate(wrong_indices[:num_show]):
        ax = axes[i]
        
        # Lấy ảnh và denormalize
        img = denormalize(test_images[idx])
        
        # Hiển thị ảnh
        ax.imshow(img)
        
        # Labels
        true_label = ds_full_eval.classes[y_true[idx]]
        pred_label = ds_full_eval.classes[y_pred[idx]]
        confidence = y_probs[idx][y_pred[idx]] * 100
        
        # Title với màu đỏ cho dự đoán sai
        title = f"True: {true_label}\nPred: {pred_label}\nConf: {confidence:.1f}%"
        ax.set_title(title, fontsize=10, color='red', fontweight='bold')
        ax.axis('off')
    
    # Ẩn các subplot thừa
    for i in range(num_show, len(axes)):
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.suptitle(f"Wrong Predictions (Showing {num_show} out of {len(wrong_indices)})", 
                 fontsize=16, fontweight='bold', y=1.02)
    plt.show()
    plt.close(fig)
    
    print(f"✓ Displayed {num_show} wrong predictions")

# Cleanup
del model, test_images
clear_cuda()
print(f"Best Fold: {best_overall_fold} with Acc: {best_overall_acc:.4f}")
```
