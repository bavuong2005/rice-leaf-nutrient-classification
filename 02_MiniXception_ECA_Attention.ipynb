```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Subset
from torchvision import transforms
from torchvision.datasets import ImageFolder
from sklearn.model_selection import StratifiedKFold, StratifiedShuffleSplit
from sklearn.metrics import classification_report
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy
import time
from PIL import Image


# ==========================================
# 1. ATTENTION MODULES
# ==========================================

class ECA(nn.Module):
    """
    Efficient Channel Attention (ECA) từ Add-ViT paper
    Formula: I' = ECA(I) = I + σ(Conv(Avg(I)))
    
    Args:
        channels: Số channels đầu vào
        kernel_size: Kích thước kernel cho 1D conv (default=3)
    """
    def __init__(self, channels, kernel_size=3):
        super().__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.conv = nn.Conv1d(1, 1, kernel_size=kernel_size, 
                              padding=(kernel_size - 1) // 2, bias=False)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Global Average Pooling: [B, C, H, W] -> [B, C, 1, 1]
        y = self.avg_pool(x)
        
        # [B, C, 1, 1] -> [B, 1, C]
        y = y.squeeze(-1).transpose(-1, -2)
        
        # 1D Convolution + Sigmoid: [B, 1, C] -> [B, 1, C]
        y = self.conv(y)
        y = self.sigmoid(y)
        
        # [B, 1, C] -> [B, C, 1, 1]
        y = y.transpose(-1, -2).unsqueeze(-1)
        
        # Residual connection
        return x + x * y.expand_as(x)


# ==========================================
# 2. MODEL ARCHITECTURE
# ==========================================

class SeparableConv2d(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3, stride=1, padding=1):
        super().__init__()
        self.depthwise = nn.Conv2d(in_channels, in_channels, kernel_size, stride, padding,
                                   groups=in_channels, bias=False)
        self.pointwise = nn.Conv2d(in_channels, out_channels, kernel_size=1, bias=False)

    def forward(self, x):
        return self.pointwise(self.depthwise(x))


class SeparableConvBlock(nn.Module):
    """
    Separable Conv Block với ECA Attention
    """
    def __init__(self, in_channels, out_channels, stride=2, use_attention=True):
        super().__init__()
        
        self.use_attention = use_attention

        # Residual path
        self.residual = nn.Conv2d(in_channels, out_channels, 1, stride, bias=False)
        self.bn_res = nn.BatchNorm2d(out_channels)

        # Main path
        self.relu1 = nn.ReLU(inplace=False)
        self.sep1 = SeparableConv2d(in_channels, out_channels)
        self.bn1 = nn.BatchNorm2d(out_channels)
        
        # ECA Attention sau separable conv đầu tiên
        if use_attention:
            self.eca1 = ECA(out_channels, kernel_size=3)

        self.relu2 = nn.ReLU(inplace=False)
        self.sep2 = SeparableConv2d(out_channels, out_channels)
        self.bn2 = nn.BatchNorm2d(out_channels)
        
        # ECA Attention sau separable conv thứ hai
        if use_attention:
            self.eca2 = ECA(out_channels, kernel_size=3)

        self.pool = nn.MaxPool2d(3, stride=stride, padding=1)

    def forward(self, x):
        # Residual path
        res = self.bn_res(self.residual(x))

        # Main path
        out = self.relu1(x)
        out = self.bn1(self.sep1(out))
        
        # Apply ECA attention
        if self.use_attention:
            out = self.eca1(out)
        
        out = self.relu2(out)
        out = self.bn2(self.sep2(out))
        
        # Apply ECA attention
        if self.use_attention:
            out = self.eca2(out)

        out = self.pool(out)
        return out + res


class XceptionECA(nn.Module):
    """
    Mini-Xception với ECA Attention
    """
    def __init__(self, num_classes=3, use_attention=True):
        super().__init__()

        self.entry = nn.Sequential(
            nn.Conv2d(3, 64, 3, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=False)
        )
        
        # ECA sau entry block
        if use_attention:
            self.entry_eca = ECA(64, kernel_size=3)
        else:
            self.entry_eca = nn.Identity()

        self.block1 = SeparableConvBlock(64, 128, use_attention=use_attention)
        self.block2 = SeparableConvBlock(128, 256, use_attention=use_attention)
        self.block3 = SeparableConvBlock(256, 512, use_attention=use_attention)

        self.exit_sep = SeparableConv2d(512, 1024)
        self.exit_bn = nn.BatchNorm2d(1024)
        self.exit_relu = nn.ReLU(inplace=False)
        
        # ECA trước global pooling
        if use_attention:
            self.exit_eca = ECA(1024, kernel_size=3)
        else:
            self.exit_eca = nn.Identity()

        self.global_pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Linear(1024, num_classes)

    def forward(self, x):
        x = self.entry(x)
        x = self.entry_eca(x)
        
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        
        x = self.exit_relu(self.exit_bn(self.exit_sep(x)))
        x = self.exit_eca(x)
        
        x = self.global_pool(x)
        x = torch.flatten(x, 1)
        return self.fc(x)


class EarlyStopping:
    def __init__(self, patience=12, min_delta=0.0):
        self.patience = patience
        self.min_delta = min_delta
        self.best_loss = None
        self.counter = 0
        self.stop = False

    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0


# ==========================================
# 3. VISUALIZATION FUNCTIONS
# ==========================================

def visualize_misclassified(model, test_loader, device, class_names, num_samples=5):
    """
    Hiển thị các ảnh bị dự đoán sai
    """
    model.eval()
    
    misclassified_images = []
    misclassified_preds = []
    misclassified_labels = []
    misclassified_indices = []
    
    batch_idx = 0
    sample_idx = 0
    
    with torch.no_grad():
        for x, y in test_loader:
            x_device = x.to(device)
            out = model(x_device)
            preds = torch.argmax(out, dim=1).cpu()
            
            # Tìm các sample bị predict sai
            for i in range(len(y)):
                if preds[i] != y[i]:
                    misclassified_images.append(x[i])
                    misclassified_preds.append(preds[i].item())
                    misclassified_labels.append(y[i].item())
                    misclassified_indices.append((batch_idx, i))
                    
                    if len(misclassified_images) >= num_samples:
                        break
                sample_idx += 1
            
            if len(misclassified_images) >= num_samples:
                break
            batch_idx += 1
    
    # Visualize
    if len(misclassified_images) > 0:
        fig, axes = plt.subplots(1, min(num_samples, len(misclassified_images)), 
                                 figsize=(4*min(num_samples, len(misclassified_images)), 4))
        
        if min(num_samples, len(misclassified_images)) == 1:
            axes = [axes]
        
        for idx, (img, pred, label) in enumerate(zip(misclassified_images[:num_samples], 
                                                       misclassified_preds[:num_samples], 
                                                       misclassified_labels[:num_samples])):
            # Denormalize image
            mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
            std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
            img = img * std + mean
            img = torch.clamp(img, 0, 1)
            
            # Convert to numpy
            img_np = img.permute(1, 2, 0).numpy()
            
            axes[idx].imshow(img_np)
            axes[idx].set_title(f'True: {class_names[label]}\nPred: {class_names[pred]}', 
                               color='red', fontsize=10)
            axes[idx].axis('off')
        
        plt.tight_layout()
        plt.show()
        
        print(f"\n📊 Found {len(misclassified_images)} misclassified samples in test set")
    else:
        print("\n🎉 No misclassified samples found! Perfect predictions!")


# ==========================================
# 4. DEVICE & AMP CONFIG
# ==========================================

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
if device == "cuda":
    torch.backends.cudnn.benchmark = True

use_amp = (device == "cuda")
if use_amp:
    scaler = torch.amp.GradScaler("cuda")
else:
    scaler = None


# ==========================================
# 5. DATA PREPARATION
# ==========================================

train_transform = transforms.Compose([
    transforms.Resize((1024, 256)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(20),
    transforms.RandomAffine(degrees=0, scale=(0.8, 1.2)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])

eval_transform = transforms.Compose([
    transforms.Resize((1024, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])

# Kaggle dataset path
DATA_DIR = '/kaggle/input/nutrientdeficiencysymptomsinrice/rice_plant_lacks_nutrients/'
CLASSES = ['Nitrogen(N)', 'Phosphorus(P)', 'Potassium(K)']
NUM_CLASSES = 3

ds_full_train = ImageFolder(DATA_DIR, transform=train_transform)
ds_full_eval = ImageFolder(DATA_DIR, transform=eval_transform)

labels = np.array(ds_full_eval.targets)

splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=42)
trainval_idx, test_idx = next(splitter.split(np.zeros(len(labels)), labels))

test_ds = Subset(ds_full_eval, test_idx)
trainval_labels = labels[trainval_idx]


# ==========================================
# 6. HYPERPARAMETERS
# ==========================================

batch_size = 32
num_epochs = 100
lr = 0.001
n_splits = 5

skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

test_loader = DataLoader(test_ds, batch_size=batch_size,
                         shuffle=False, num_workers=2, pin_memory=True)

results = []
best_overall_acc = 0
best_overall_state = None
best_overall_fold = None

# Print model info
model = XceptionECA(num_classes=NUM_CLASSES, use_attention=True)
print("\n" + "="*50)
print("MODEL ARCHITECTURE: Xception with ECA Attention")
print("="*50)
print(model)
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f'\nTotal parameters: {total_params:,} ({total_params/1e6:.2f}M)')
print(f'Trainable parameters: {trainable_params:,} ({trainable_params/1e6:.2f}M)')
print(f'Dataset: {len(ds_full_train)} images')
print(f'Classes: {CLASSES}')
print("="*50 + "\n")


# ==========================================
# 7. K-FOLD TRAINING LOOP
# ==========================================

fold = 1
total_training_time = 0

for train_idx, val_idx in skf.split(np.zeros(len(trainval_labels)), trainval_labels):
    print(f"\n{'='*15} FOLD {fold}/{n_splits} {'='*15}")

    train_subset = Subset(ds_full_train, trainval_idx[train_idx])
    val_subset   = Subset(ds_full_eval, trainval_idx[val_idx])

    train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)
    val_loader   = DataLoader(val_subset, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)

    model = XceptionECA(num_classes=NUM_CLASSES, use_attention=True).to(device)
    
    opt = optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(opt, mode='min', factor=0.1, patience=5)
    
    criterion = nn.CrossEntropyLoss()
    earlystop = EarlyStopping(patience=12)

    train_losses, val_losses = [], []
    train_accs, val_accs = [], []
    epoch_times = []
    train_times = []
    val_times = []
    best_val_loss = float("inf")
    best_state = None

    fold_start_time = time.time()

    for epoch in range(num_epochs):
        epoch_start_time = time.time()
        
        # ==================== TRAIN ====================
        train_start = time.time()
        model.train()
        running_loss = 0
        correct = 0
        total = 0
        
        for x, y in train_loader:
            x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
            
            opt.zero_grad()
            
            if use_amp:
                with torch.amp.autocast("cuda"):
                    out = model(x)
                    loss = criterion(out, y)
                scaler.scale(loss).backward()
                scaler.step(opt)
                scaler.update()
            else:
                out = model(x)
                loss = criterion(out, y)
                loss.backward()
                opt.step()
            
            running_loss += loss.item()
            preds = torch.argmax(out, dim=1)
            correct += (preds == y).sum().item()
            total += y.size(0)
            
        train_loss = running_loss / len(train_loader)
        train_acc = correct / total
        train_time = time.time() - train_start
        
        # ==================== VALIDATION ====================
        val_start = time.time()
        model.eval()
        val_loss_sum = 0
        val_correct = 0
        val_total = 0
        
        with torch.no_grad():
            for x, y in val_loader:
                x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)

                if use_amp:
                    with torch.amp.autocast("cuda"):
                        out = model(x)
                        loss = criterion(out, y)
                else:
                    out = model(x)
                    loss = criterion(out, y)
                
                val_loss_sum += loss.item()
                preds = torch.argmax(out, dim=1)
                val_correct += (preds == y).sum().item()
                val_total += y.size(0)
        
        val_loss = val_loss_sum / len(val_loader)
        val_acc = val_correct / val_total
        val_time = time.time() - val_start
        
        # ==================== TIMING ====================
        epoch_time = time.time() - epoch_start_time
        epoch_times.append(epoch_time)
        train_times.append(train_time)
        val_times.append(val_time)
        
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)
        
        # Print với timing chi tiết
        print(f"Epoch {epoch+1}/{num_epochs} | "
              f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
              f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} | "
              f"Time: {epoch_time:.2f}s (Train: {train_time:.2f}s, Val: {val_time:.2f}s, "
              f"Val%: {100*val_time/epoch_time:.1f}%)")
        
        scheduler.step(val_loss)

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_state = deepcopy(model.state_dict())
            
        earlystop(val_loss)
        if earlystop.stop:
            print(f"Early stopping at epoch {epoch+1}")
            break

    fold_time = time.time() - fold_start_time
    total_training_time += fold_time
    avg_epoch_time = np.mean(epoch_times)
    avg_train_time = np.mean(train_times)
    avg_val_time = np.mean(val_times)
    val_percentage = 100 * avg_val_time / avg_epoch_time
    
    print(f"\n⏱️  Fold {fold} Training Time: {fold_time:.2f}s ({fold_time/60:.2f}min)")
    print(f"⏱️  Average Epoch Time: {avg_epoch_time:.2f}s")
    print(f"⏱️  Average Train Time: {avg_train_time:.2f}s ({100*avg_train_time/avg_epoch_time:.1f}%)")
    print(f"⏱️  Average Val Time: {avg_val_time:.2f}s ({val_percentage:.1f}%)")

    # PLOT CURVES
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label="Train")
    plt.plot(val_losses, label="Val")
    plt.title(f"Fold {fold} Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label="Train")
    plt.plot(val_accs, label="Val")
    plt.title(f"Fold {fold} Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()

    # TEST EVALUATION
    model.load_state_dict(best_state)
    model.eval()
    
    y_true, y_pred = [], []
    with torch.no_grad():
        for x, y in test_loader:
            x = x.to(device)
            if use_amp:
                with torch.amp.autocast("cuda"):
                    out = model(x)
            else:
                out = model(x)
            pred = torch.argmax(out, dim=1).cpu().numpy()
            y_pred.extend(pred)
            y_true.extend(y.numpy())
            
    acc = np.mean(np.array(y_true) == np.array(y_pred))
    print(f"\n>>> Test Accuracy Fold {fold}: {acc:.4f}")
    print(classification_report(y_true, y_pred, target_names=CLASSES))
    
    # VISUALIZE MISCLASSIFIED SAMPLES
    print(f"\n{'='*20} MISCLASSIFIED SAMPLES {'='*20}")
    visualize_misclassified(model, test_loader, device, CLASSES, num_samples=5)
    
    results.append(acc)
    
    if acc > best_overall_acc:
        best_overall_acc = acc
        best_overall_state = best_state
        best_overall_fold = fold
        
    fold += 1


# ==========================================
# 8. FINAL SUMMARY
# ==========================================

torch.save(best_overall_state, "best_xception_eca.pth")
avg_acc = np.mean(results)
miss_rate = 1 - avg_acc

model_params = sum(
    p.numel() for p in XceptionECA(num_classes=NUM_CLASSES).parameters()
    if p.requires_grad
)

print("\n" + "="*20 + " FINAL SUMMARY " + "="*20)
print(f"Average 5-Fold Accuracy: {avg_acc:.4f}")
print(f"Miss-classification Rate: {miss_rate:.4f}")
print(f"Trainable Parameters: {model_params:,}  ({model_params/1e6:.2f}M)")
print(f"Best Fold: {best_overall_fold} with Accuracy: {best_overall_acc:.4f}")
print(f"Total Training Time: {total_training_time:.2f}s ({total_training_time/60:.2f}min)")
print(f"Average Time per Fold: {total_training_time/n_splits:.2f}s ({total_training_time/n_splits/60:.2f}min)")
print("Best Model Saved → best_xception_eca.pth")
print("="*56)
```
